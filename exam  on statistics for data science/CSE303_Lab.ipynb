```python
def mean_md_saful(l):
  sum = 0
  for i in l:
    sum = sum + i;
  mean = sum/ len(l)
  return mean
```

```python
def selectionSort(l):
    ln = int(len(l))
    for i in range(ln):
        min_index = i
        for j in range(i + 1, ln):
            if l[j] < l[min_index]:
                min_index = j
        #swap a,b = b,a
        (l[i], l[min_index]) = (l[min_index], l[i])
```

```python
def median_md_saful(l):
  selectionSort(l)
  ln = (len(l))
  if ln%2!=0:
    median= l[int((ln+1)/2)]
  else:
    median= l[int(int(ln/2 +(ln+1)/2)/2)]
  return median
```

```python
def varience_md_saiful(l):
  mean = mean_md_saful(l)
  sum = 0
  for i in range(0,len(l),1):
    sum+=(l[i]-mean)**2
  return (sum/(len(l)-1))
```
