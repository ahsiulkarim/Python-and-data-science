3. **Generate a correlation heatmap among all attributes. Find the top 5 positively correlated attribute pairs and top 5 negatively correlated attribute pairs.**

```python
# calulating correlation matrix
corr = df.corr()

# creating heatmap
plt.figure(figsize=(5,6))
heatmap = plt.imshow(corr, cmap='Blues')
plt.colorbar(heatmap)   

corr_pairs = corr.unstack()

sorted_pairs = corr_pairs.sort_values()

top_positive = sorted_pairs.tail(5)
top_negative = sorted_pairs.head(5)

print("Top 5 negative correlations:\n",top_negative)
print("\nTop 5 positive correlations:\n" , top_positive)
```
